```python
class TreeNode:
    def __init__(self, value: int, left = None, right = None):
        self.value = value
        self.left = left
        self.right = right

def preorder(root):
    if not root:
        return []
    return [root.value] + preorder(root.left) + preorder(root.right)

def inorder(root):
    if not root:
        return []
    return inorder(root.left) + [root.value] + inorder(root.right)

def postorder(root):
    if not root:
        return []
    return postorder(root.left) + postorder(root.right) + [root.value]

def preorder_iter(root):
    if not root:
        return []
    stack = [root]
    out = []
    while stack:
        node = stack.pop()
        out.append(node.value)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return out

def inorder_iter(root):
    stack = []
    current = root
    out = []

    while stack or current:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        out.append(current.value)
        current = current.right
    return out
```

```python
def _is_bst_helper(root, prev):

    if root is None:
        return True
    
    if not _is_bst_helper(root.left, prev):
        return False
   
    if prev[0] >= root.value:
        return False
    
    prev[0] = root.value
    return _is_bst_helper(root.right, prev)

def is_bst(root):
    prev = [float('-inf')]
    isbst_bool = _is_bst_helper(root, prev)
    return isbst_bool, prev

root = TreeNode(12)
root.left = TreeNode(7)
root.right = TreeNode(20)
root.left.left = TreeNode(3)
root.left.right = TreeNode(10)

root.right.left = TreeNode(11)
root.right.right = TreeNode(25)

root = TreeNode(2)
root.left = TreeNode(1)
root.right = TreeNode(4)
root.left.right = TreeNode(3)
is_bst(root)
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.height = 0
        self.left = None
        self.right = None
    
class AVL:
    def __init__(self):
        self.root = None
    
    def height(self, node):
        return node.height if node else 0

    def balance(self, node):
        return self.height(node.left) - self.height(node.right)
    
    def right_rotate(self, y):
        x = y.left
        t2 = x.right

        x.right = y
        y.left = t2

        y.height = 1 + max(self.height(y.left), self.height(y.right))
        x.height = 1 + max(self.height(x.left), self.height(x.right))

        return x
    
    def left_rotate(self, x):
        y = x.right
        t2 = y.left

        y.left = x
        x.right = t2

        x.height = 1 + max(self.height(x.left), self.height(x.right))
        y.height = 1 + max(self.height(y.left), self.height(y.right))

        return y
    
    def print_tree(self):
        self._print_tree(self.root, "Root")

    def _print_tree(self, node, label):
        if node is None:
            return
        
        print(f"{label}: {node.value}")
        
        if node.left or node.right:
            if node.left:
                self._print_tree(node.left, f"  L of {node.value}")
            else:
                print(f"  L of {node.value}: None")
                
            if node.right:
                self._print_tree(node.right, f"  R of {node.value}")
            else:
                print(f"  R of {node.value}: None")

    def _insert(self, root, value):
        if root is None:
            return TreeNode(value)

        if value < root.value:
            root.left = self._insert(root.left, value)
        elif value > root.value:
            root.right = self._insert(root.right, value)
        else:
            return root
        
        root.height = 1 + max(self.height(root.left), self.height(root.right))
    
        bf = self.balance(root)
        # LL
        if bf > 1 and value < root.left.value:
            return self.right_rotate(root)

        # RR
        if bf < -1 and value > root.right.value:
            return self.left_rotate(root)

        # LR
        if bf > 1 and value > root.left.value:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)

        # RL
        if bf < -1 and value < root.right.value:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)

        return root

    def insert(self, value):
        self.root = self._insert(self.root, value)


    
mytree = AVL()
for item in [3, 7, 12, 20]:
    mytree.insert(item)
mytree.print_tree()
```
