```python
class Node:
    def __init__(self, value):
        self.data = value 
        self.next: Node | None = None


n1 = Node(16)
n2 = Node(22)
n3 = Node(83)
n4 = Node(49)
head = n1
head.next = n2
head.next.next = n3
head.next.next.next = n4
```

```python
curr = head
count = 0


while curr != None:
    if count == 2:
        print(curr.data)
        break
    count = count + 1
    curr = curr.next


print(id(n3))  
```
